# PANDAS PRACTICE ON sales_data_sample.csv FROM KAGGLE (2800+ records)

```python
import pandas as pd
df= pd.read_csv(r"sales_data_sample.csv" , encoding='ISO-8859-1')
```

# Select only the columns: Order no, quantity ordered, price of each

```python
df[['ORDERNUMBER','QUANTITYORDERED','PRICEEACH']]
```

# Create a new column called TOTAL_SALES which is the quantity multiplied by price (QUANTITYORDERED * PRICEEACH).

```python
df.columns
```

```python
df['TOTAL_SALES']= df['QUANTITYORDERED']* df['PRICEEACH']
```

```python
df.columns
```

# Find the total sales per order (ORDERNUMBER).

```python
df.groupby('ORDERNUMBER')['TOTAL_SALES'].sum()
```

# Find the top 5 orders with the highest total sales.

```python
df.sort_values('TOTAL_SALES', ascending= False).head()
```

# Find the total sales per product line (PRODUCTLINE).

```python
df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
```

# Find the average sale amount (SALES) per country.

```python
df.columns
```

```python
df.groupby('COUNTRY')['SALES'].mean().head()
```

# Find all orders where TOTAL_SALES is greater than 5000.

```python
df[df['TOTAL_SALES']>5000].head(2)
```

# Create a new column called SALES_CATEGORY:

"High" if TOTAL_SALES > 3000

"Medium" if TOTAL_SALES between 1000 and 3000

"Low" if TOTAL_SALES < 1000

```python
import numpy as np

df['SALES_CATEGORY'] = np.where(df['TOTAL_SALES'] > 3000, 'High',
                        np.where(df['TOTAL_SALES'] >= 1000, 'Medium', 'Low'))

df[['TOTAL_SALES','SALES_CATEGORY']].head(10)
```
